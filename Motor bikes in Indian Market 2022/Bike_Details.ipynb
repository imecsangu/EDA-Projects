```python
def plot_continious(df, column):
    plt.figure(figsize = (10,6))
    plt.subplot(1,3,1)
    sns.histplot(df[column], color = 'orange')
    plt.subplot(1,3,2)
    sns.boxplot(data = df, x = column, color = 'orange')
    plt.subplot(1,3,3)
    sns.kdeplot(data = df, x = column)
    plt.show()
    
plot_continious(df[df['Electric']!=1], 'Price')
```

**1.What are the top 10 brands with most bikes in indian market** 

```python
Top10 = df['Brand'].value_counts().head(10)
Top10
```

```python
plt.figure(figsize = (11,6))
sns.countplot(x = df.loc[df['Brand'].isin(Top10.index), 'Brand'])
plt.ylim(5, 35)
plt.xlabel('Brands with most bike models in india')
plt.ylabel('Number of bikes')
plt.title('Top 10 Brands with most indian bikes')
plt.show()
```

**The above count plot says Hero, Honda, Yamaha, Bajaj, Kawasaki, TVS, BMW, Hearly-Davidson,Triumph, Ducati are top 10 brands in indian market**

**2.Which brand produces the bikes with best mileage in india(Non Electric).?**

```python
Top_Mileage = df[df['Electric']==0].groupby(['Brand'])['Mileage'].mean().nlargest(5)
Top_Mileage
```

```python
plt.figure(figsize = (11,6))
sns.barplot(x = Top_Mileage.values, y = Top_Mileage.index, orient = 'h')
plt.xlabel('Brands with top mileage')
plt.ylabel('Brands')
plt.title('Top 5 brands with top mileage')
plt.show()
```

**Yezdi and Hero are the top bikes in india to produce great mileage**
