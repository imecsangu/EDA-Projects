#### Task - 2 Add month column

```python
df['Month'] = df['Order Date'].str[0:2]
df.head()
```

```python
nan_df = df[df.isna().any(axis = 1)]
display(nan_df)

df = df.dropna(how = 'all')
df.head()
```

##### Find 'Or' and delete it

```python
df = df[df['Order Date'].str[0:2]!='Or']
```

```python
df['Month'] = df['Month'].astype(int)
```

#### Convert columns to the correct type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered']) # make int
df['Price Each'] = pd.to_numeric(df['Price Each']) # make float
```

#### Task - 3 add sales column

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each'] # sales = Quantity ordered * Price each
df.head()
```

#### Add a city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
    
# df['City'] = df['Purchase Address'].apply(lambda x : x.split(',')[1])
# df['City'] = df['Purchase Address'].apply(lambda x : get_city(x) + ' ' + get_state(x))
df['City'] = df['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")
df.head()
```

#### Question 3: What time should we display advertisements to maximize likelyhood of customers buying products.?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df.head()
```

```python
results = df.groupby('Hour').count()
results
```

```python
hours = [hour for hour, df in df.groupby('Hour')]

plt.plot(hours, results['Sales'])
plt.xticks(hours, size = 8)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

The line graph shows and tell us that the time for showing advertisements to maxmize customer likelihood buying products is morning in between 10-11 AM and in the evening time between 18-19 PM 

#### Question 4: What products are most often sold together.?

```python
new_df = df[df['Order ID'].duplicated(keep = False)]
new_df.head()
new_df['Grouped'] = new_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
new_df.head()
new_df = new_df[['Order ID', 'Grouped']].drop_duplicates()
new_df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

By seeing new_df result is is saying most sold items together is iPhone, Lightning Charging Cable with an average mean of 1005
