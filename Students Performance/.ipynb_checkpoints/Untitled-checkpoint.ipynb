```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('exams.csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data = df, x = 'math score', kde = True, bins = 30, legend = True)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data = df, x = 'reading score', kde = True, bins = 30, legend = True)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data = df, x = 'writing score', kde = True, bins = 30, legend = True)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x = 'race/ethnicity', data = df, palette = 'pastel')
plt.xticks(rotation = 45)
plt.xlabel('race/ethnicity')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x = 'parental level of education', data = df, palette = 'pastel')
plt.xticks(rotation = 45)
plt.xlabel('race/ethnicity')
plt.show()
```

```python
label = df['parental level of education'].value_counts().index
```

```python
label
```

```python
sizes = df['parental level of education'].value_counts().values
```

```python
sizes
```

```python
# colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize=(10,8))
plt.pie(labels = label, x = sizes, autopct = '%1.1f%%')
plt.title('parental lavel of education in pie chart')
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.countplot(x=df["race/ethnicity"], hue=df["gender"], palette="colorblind")
plt.xlabel("Race/Ethnicity in countplot")
plt.ylabel("Frequency")
plt.show()
```
